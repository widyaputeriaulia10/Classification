```python
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize  
from nltk.tokenize import sent_tokenize 
from nltk.corpus import stopwords #import stopwords
from nltk.tokenize import RegexpTokenizer
import re #import regular expression
from sklearn.feature_extraction.text import TfidfVectorizer
from textblob import TextBlob
from textblob.classifiers import NaiveBayesClassifier
from sklearn.naive_bayes import MultinomialNB
```

```python
data = pd.read_excel("sentimentanalysis.xlsx")
datatest = pd.read_excel("testing.xlsx")
text=data['text']
sentiment= data['sentiment']
textset = datatest['text']
senset = datatest['sentiment']
```

```python
def clean(tweet): 
        smtr = ' '.join(re.sub(r'@[A-Za-z0-9]+','',tweet).split()) 
        smtr = ' '.join(re.sub('https?://[A-Za-z0-9./]+','',smtr).split())
        smtr = ' '.join(re.sub("[^a-zA-Z]", " ", smtr).split())
        smtr = ' '.join(re.sub("RT", " ", smtr).split())
        smtr = smtr.lower().split()
        return smtr

cleantweet = []
i = 0
while i < len(text):
    baru = clean(text[i])
    cleantweet.append(baru)
    i += 1
    
print(cleantweet)
```

```python
stopword = ['a','about','above','add','after','again','against','all','am','amp','an',
                 'and','any','are','arent','as','at','be','because','been',
                 'before','being','below','between','both','but','by','cant','cannot',
                 'could','couldnt','did','didnt','didn','do','does','doesnt','doing',
                 'down','during','each','few','for','from','further','had','hadnt','has','have','having',
                 'he','her','here','hers','herself','him','himself','his','how','i','if','in','into','is','it',
                 'its','itself','m','me','more','most','must','my','myself','nd','no','nor','of','off','on','once','only','or',
                 'other','ought','our','ours','ourselves','out','over','own','said','same','she','should',
                 'so','some','such','t','than','that','the','their','theirs','them','themselves','then','there',
                 'these','they','this','those','through','to','too',
                 'under','until','up','ve','va','very','was','we','were','what','when','where','which','while','who','whom','why','with',
                 'would','you','your','yours','yourself','yourselves']
sw=[]

for conten in cleantweet:
    kata= list(filter(lambda x: x not in stopword, conten))
    kata = ' '.join(kata)
    sw.append(kata)

data['stopword'] = sw

print(sw)
```

```python
dicti = []
j = 0
while j < len(text):
    train = (sw[j], sentiment[j])
    dicti.append(train)
    
    j += 1

print(dicti)
```

```python
cl = NaiveBayesClassifier(dicti)
#cl.classify("I feel amazing!")
blob = TextBlob("I hate this", classifier=cl)
print(blob)
print(blob.classify())
```
